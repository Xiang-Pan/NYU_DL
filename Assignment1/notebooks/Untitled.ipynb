# Activation

```python
import torch.nn as nn
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## ReLU

```python
activation_fun = nn.ReLU()
```

```python
x_list = np.linspace(-10, 10, 1000)
y_list = activation_fun(torch.Tensor(x_list))
dy_list = []
for x in x_list:
    x_tensor = torch.Tensor([x])
    x_tensor.requires_grad_(True)

    y = activation_fun(x_tensor)
    y.backward()
    dy_list.append(float(x_tensor.grad[0]))

plt.figure(figsize=(10, 5))
plt.plot(x_list, y_list)
plt.plot(x_list, dy_list)
plt.legend(['ReLU','Derivative for ReLU'])
plt.savefig("../imgs/ReLU")
```

## LeakyReLU

```python
import torch
activation_fun = nn.LeakyReLU()
```

```python
x_list = np.linspace(-10, 10, 1000)
y_list = activation_fun(torch.Tensor(x_list))
dy_list = []
for x in x_list:
    x_tensor = torch.Tensor([x])
    x_tensor.requires_grad_(True)

    y = activation_fun(x_tensor)
    y.backward()
    dy_list.append(float(x_tensor.grad[0]))

plt.figure(figsize=(10, 5))
plt.plot(x_list, y_list)
plt.plot(x_list, dy_list)
plt.title("LeakyReLU")
plt.legend(['LeakyReLU(negative_slope=0.01)','Derivative for LeakyReLU(negative_slope=0.01)'])
plt.savefig("../imgs/LeakyReLU")
```

```python
x_list = np.linspace(-10, 10, 1000)
y_list = activation_fun(torch.Tensor(x_list))
dy_list = []
for x in x_list:
    x_tensor = torch.Tensor([x])
    x_tensor.requires_grad_(True)

    y = activation_fun(x_tensor)
    y.backward()
    dy_list.append(float(x_tensor.grad[0]))

plt.figure(figsize=(10, 5))
plt.xlim([-10, 1])
plt.ylim([-0.02, 0.02])
plt.plot(x_list, y_list)
plt.plot(x_list, dy_list)
plt.title("LeakyReLU_zoomin")
plt.legend(['LeakyReLU(negative_slope=0.01)','Derivative for LeakyReLU(negative_slope=0.01)'])
plt.savefig("../imgs/LeakyReLU_zoomin")
```

## softplus

```python
activation_fun = nn.Softplus(beta=1, threshold=20)
```

```python
x_list = np.linspace(-10, 10, 1000)
y_list = activation_fun(torch.Tensor(x_list))
# x_l = np.linspace(-10, 0, 1000)
# x_r = np.linspace(0, 1, 1000)
dy_list = []
for x in x_list:
    x_tensor = torch.Tensor([x])
    x_tensor.requires_grad_(True)

    y = activation_fun(x_tensor)
    y.backward()
    dy_list.append(float(x_tensor.grad[0]))
plt.figure(figsize=(10, 5))
plt.xlim([-10, 10])
plt.plot(x_list, y_list)
plt.plot(x_list, dy_list)
plt.title("Softplus")
plt.legend(['Softplus(beta=1, threshold=20)','Derivative for Softplus(beta=1, threshold=20)'])
plt.savefig("../imgs/Softplus")
```
